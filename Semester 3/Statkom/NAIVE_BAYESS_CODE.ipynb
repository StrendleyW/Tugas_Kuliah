# DATA TRAINING

Mengambil 70 persen dari class Y dan class N

Dalam contoh ini head() digunakan untuk mengambil sejumlah baris teratas dari DataFrame yang sesuai dengan persentase yang diinginkan. Misalnya, df_Y.head(num_rows_to_take_Y) akan mengambil sejumlah baris teratas dari DataFrame df_Y yang setara dengan 70% dari total baris DataFrame tersebut. concat() digunakan untuk menggabungkan class Y dan class N yang telah diambil 70% nya.

```python
# Filter data yang memiliki nilai 'Y' pada kolom 'Class'
df_Y = df[df['Sesuai'] == 'Y']
df_N = df[df['Sesuai'] == 'N']

# Menghitung jumlah baris yang akan diambil (70%)
percentage = 0.7
num_rows_to_take_Y = round(len(df_Y) * percentage)
num_rows_to_take_N = round(len(df_N) * percentage)

# Mengambil 70% data yang memiliki nilai 'Y' secara berurutan
df_Y_sampled = df_Y.head(num_rows_to_take_Y)
df_N_sampled = df_N.head(num_rows_to_take_N)

# Menggabungkan hasil sampling dari kedua kelompok
final_sampled_data = pd.concat([df_Y_sampled, df_N_sampled]).reset_index(drop=True)

# Memulai indeks dari 1
final_sampled_data.index += 1

# Output Hasil
print(final_sampled_data.to_string(max_rows=None))
```

# Kode Naive Bayess

```python
def naive_bayes_prediction(tinggi_badan, berat_badan, sex, lingkat_pergelangan ,lingkar_perut, statusgizi_riil, statusgizi_nby):
    persentase_tinggi_badan = {'Rendah': {'Y': prob_tinggi_rendah_Y, 'N': prob_tinggi_rendah_N},
                            'Sedang': {'Y': prob_tinggi_sedang_Y, 'N': prob_tinggi_sedang_N},
                            'Tinggi': {'Y': prob_tinggi_tinggi_Y, 'N': prob_tinggi_tinggi_N}}

    persentase_berat_badan = {'Ringan': {'Y': prob_berat_ringan_Y, 'N': prob_berat_ringan_N},
                            'Sedang': {'Y': prob_berat_sedang_Y, 'N': prob_berat_sedang_N},
                                'Berat': {'Y': prob_berat_berat_Y, 'N': prob_berat_berat_N}}
    
    persentase_sex = {'L': {'Y': prob_sex_L_Y, 'N': prob_sex_L_N},
                            'P': {'Y': prob_sex_P_Y, 'N': prob_sex_P_N}}

    persentase_lingkar_pergelangan = {'S': {'Y': prob_gelang_s_Y, 'N': prob_gelang_s_N},
                            'M': {'Y': prob_gelang_m_Y, 'N': prob_gelang_m_N},
                                'L': {'Y': prob_gelang_l_Y, 'N': prob_gelang_l_N},
                                'XL': {'Y': prob_gelang_xl_Y, 'N': prob_gelang_xl_N},
                                'XLL': {'Y': prob_gelang_xxl_Y, 'N': prob_gelang_xxl_N}}
    
    persentase_lingkar_perut = {'S': {'Y': prob_perut_s_Y, 'N': prob_perut_s_N},
                            'M': {'Y': prob_perut_m_Y, 'N': prob_perut_m_N},
                                'L': {'Y': prob_perut_l_Y, 'N': prob_perut_l_N},
                                'XL': {'Y': prob_perut_xl_Y, 'N': prob_perut_xl_N},
                                'XLL': {'Y': prob_perut_xxl_Y, 'N': prob_perut_xxl_N}}
    
    persentase_statusgizi_riil = {'Kurang': {'Y': prob_giziriil_kurang_Y, 'N': prob_giziriil_kurang_N},
                            'Normal': {'Y': prob_giziriil_normal_Y, 'N': prob_giziriil_normal_N},
                                'Obese Ringan': {'Y': prob_giziriil_obeseringan_Y, 'N': prob_giziriil_obeseringan_N},
                                'Obese Sedang': {'Y': prob_giziriil_obesesedang_Y, 'N': prob_giziriil_obesesedang_N},
                                'Obese Berat': {'Y': prob_giziriil_obeseberat_Y, 'N': prob_giziriil_obeseberat_N}}

    persentase_statusgizi_nbc = {'Kurang': {'Y': prob_gizinbc_kurang_Y, 'N': prob_gizinbc_kurang_N},
                            'Normal': {'Y': prob_gizinbc_normal_Y, 'N': prob_gizinbc_normal_N},
                                'Obese Ringan': {'Y': prob_gizinbc_obeseringan_Y, 'N': prob_gizinbc_obeseringan_N},
                                'Obese Sedang': {'Y': prob_gizinbc_obesesedang_Y, 'N': prob_gizinbc_obesesedang_N},
                                'Obese Berat': {'Y': prob_gizinbc_obeseberat_Y, 'N': prob_gizinbc_obeseberat_N}}
    
    # Menghitung prediksi untuk kelas 'Tepat'
    prediksi_Y = (
        prob_Y *
        persentase_tinggi_badan.get(tinggi_badan, 0).get('Y', 0) *
        persentase_berat_badan.get(berat_badan, 0).get('Y', 0) *
        persentase_sex.get(sex, 0).get('Y', 0) *
        persentase_lingkar_pergelangan.get(lingkat_pergelangan, 0).get('Y', 0) *
        persentase_lingkar_perut.get(lingkar_perut, 0).get('Y', 0) *
        persentase_statusgizi_riil.get(statusgizi_riil, 0).get('Y', 0) *
        persentase_statusgizi_nbc.get(statusgizi_nby, 0).get('Y', 0)
    )

    # Menghitung prediksi untuk kelas 'Terlambat'
    prediksi_N = (
        prob_Y *
        persentase_tinggi_badan.get(tinggi_badan, 0).get('N', 0) *
        persentase_berat_badan.get(berat_badan, 0).get('N', 0) *
        persentase_sex.get(sex, 0).get('N', 0) *
        persentase_lingkar_pergelangan.get(lingkat_pergelangan, 0).get('N', 0) *
        persentase_lingkar_perut.get(lingkar_perut, 0).get('N', 0) *
        persentase_statusgizi_riil.get(statusgizi_riil, 0).get('N', 0) *
        persentase_statusgizi_nbc.get(statusgizi_nby, 0).get('N', 0)
    )

    print(f'Persentase Y: {prediksi_Y}')
    print(f'Persentase N: {prediksi_N}')

    prediksi = max(prediksi_Y, prediksi_N)
    if prediksi == prediksi_Y:
        return 'Y'
    else:
        return 'N'
    

    
tinggi_badan = input('Masukkan kategori tinggi anda! (Rendah/Sedang/Tinggi): ')
berat_badan = input('Masukkan kategori berat anda (Ringan/Sedang/Berat: ')
sex = input('Jenis Kelamin? (L/P): ')
lingkar_pergelangan = input('Masukkan kategori lingkar pergelangan (S/M/L/XL/XXL): ')
lingkar_perut = input('Masukkan kategori lingkar perut (S/M/L/XL/XXL): ')
statusgizi_riil = input('Masukkan status gizi (perhitungan riil) (Kurang/Normal/Obese Ringan/Obese Sedang/Obese Berat): ')
statusgizi_nby = input('Masukkan status gizi (perhitungan NBY) (Kurang/Normal/Obese Ringan/Obese Sedang/Obese Berat): ')



kelas_prediksi = naive_bayes_prediction(tinggi_badan, berat_badan, sex, lingkar_pergelangan, lingkar_perut, statusgizi_riil, statusgizi_nby)
print("Prediksi Kelas: ", kelas_prediksi)
```

Mengambil 30 persen dari data class Y dan class N

Dalam contoh ini, kita menggunakan metode tail() untuk mendapatkan sisa data yang tidak termasuk dalam 70%. Selanjutnya, kita menggunakan pendekatan yang serupa untuk mengambil 30% dari sisa data tersebut. 

```python
# Mengambil sisa data (30%)
df_remaining_Y = df_Y.tail(len(df_Y) - num_rows_to_take_Y)
df_remaining_N = df_N.tail(len(df_N) - num_rows_to_take_N)

# Menghitung jumlah baris yang akan diambil dari sisa data (30%)
num_rows_to_take_remaining_Y = int(len(df_remaining_Y))
num_rows_to_take_remaining_N = int(len(df_remaining_N))

# Mengambil 30% sisa data secara berurutan
df_remaining_sampled_Y = df_remaining_Y.head(num_rows_to_take_remaining_Y).reset_index(drop=True)
df_remaining_sampled_N = df_remaining_N.head(num_rows_to_take_remaining_N).reset_index(drop=True)

# Menggabungkan hasil sampling dari kedua kelompok
final_sampled_remaining_data = pd.concat([df_remaining_sampled_Y, df_remaining_sampled_N]).reset_index(drop=True)

# Memulai indeks dari 1
final_sampled_remaining_data.index += 1

# Menampilkan DataFrame hasil sampling
print(final_sampled_remaining_data.to_string(max_rows=None))
```
